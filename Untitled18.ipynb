```python
# train_ppo_minigrid.py
import gymnasium as gym
import numpy as np
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv, VecTransposeImage
from stable_baselines3.common.monitor import Monitor
from minigrid.wrappers import RGBImgObsWrapper, ImgObsWrapper
import os

ENV_ID = "MiniGrid-Empty-8x8-v0"
MODEL_DIR = "./models"
os.makedirs(MODEL_DIR, exist_ok=True)

def make_env():

    env = gym.make(ENV_ID)
    # Trả về ảnh RGB (H, W, 3)
    env = RGBImgObsWrapper(env)  # trả obs dưới dạng ảnh RGB
    env = ImgObsWrapper(env)     # observation key chỉ còn image
    env = Monitor(env)           # ghi reward / episode length cho SB3 logging
    return env

# SB3 expects vectorized env. DummyVecEnv wraps callable that returns env.
vec_env = DummyVecEnv([make_env])
# VecTransposeImage converts obs shape (H, W, C) -> (C, H, W) for CNN policies
vec_env = VecTransposeImage(vec_env)

print("Observation space:", vec_env.observation_space)
print("Action space:", vec_env.action_space)
```

```python
from stable_baselines3.common.callbacks import CheckpointCallback

# hyperparams
TIMESTEPS = 200_000    # điều chỉnh tuỳ máy; tăng nếu muốn expert mạnh hơn
POLICY = "CnnPolicy"
MODEL_PATH = os.path.join(MODEL_DIR, "ppo_minigrid")

# checkpoint để lưu model tạm khi huấn luyện lâu
checkpoint_cb = CheckpointCallback(save_freq=50_000, save_path=MODEL_DIR,
                                   name_prefix="ppo_minigrid_chk")

model = PPO(POLICY, vec_env, verbose=1, tensorboard_log="./tb_ppo_minigrid/")
model.learn(total_timesteps=TIMESTEPS, callback=checkpoint_cb)
model.save(MODEL_PATH)
print("Saved PPO model to", MODEL_PATH)
```

```python
import pickle
from tqdm import trange
import gymnasium as gym

# Load model (nếu chạy trong file khác)
# model = PPO.load(MODEL_PATH)

# Tạo môi trường “đơn” (không vector) để thu thập và dễ dàng lưu obs
def make_single_env():
    env = gym.make(ENV_ID)
    env = RGBImgObsWrapper(env)
    env = ImgObsWrapper(env)
    return env

env = make_single_env()

dataset = []
NUM_EP = 200  # số episode expert muốn thu thập

for ep in trange(NUM_EP):
    obs, info = env.reset()
    done = False
    steps = 0
    while not done:
        # model expects transposed image (C,H,W) internally, but model.predict accepts
        # the raw observation when using non-vector env, stable-baselines3 handles shapes.
        action, _ = model.predict(obs, deterministic=True)
        new_obs, reward, terminated, truncated, info = env.step(int(action))
        done = terminated or truncated

        # Lưu observation (copy để tránh reference), action và reward
        sample = {
            "obs": np.array(obs),   # H x W x C
            "action": int(action),
            "reward": float(reward),
            "done": bool(done)
        }
        dataset.append(sample)

        obs = new_obs
        steps += 1

    # (tuỳ chọn) in/ghi log ngắn
    if (ep + 1) % 50 == 0:
        print(f"Collected {ep+1} / {NUM_EP} episodes")

# Lưu dataset
with open("expert_ppo.pkl", "wb") as f:
    pickle.dump(dataset, f)

print("Saved expert_ppo.pkl with", len(dataset), "timesteps from", NUM_EP, "episodes")
```

```python
import pickle
import numpy as np
import random
from copy import deepcopy

with open("expert_ppo.pkl", "rb") as f:
    expert_data = pickle.load(f)

print("Loaded expert data:", len(expert_data), "steps")
```

```python
def add_action_flip_noise(data, p=0.1, action_space=7):
    noisy = deepcopy(data)
    for sample in noisy:
        if random.random() < p:
            current = sample["action"]
            other_actions = [a for a in range(action_space) if a != current]
            sample["action"] = random.choice(other_actions)
    return noisy
```

```python
def add_random_action_noise(data, p=0.1, action_space=7):
    noisy = deepcopy(data)
    for sample in noisy:
        if random.random() < p:
            sample["action"] = random.randint(0, action_space - 1)
    return noisy
```

```python
def add_gaussian_noise(data, std=10):
    noisy = deepcopy(data)
    for sample in noisy:
        img = sample["obs"].astype(np.float32)
        noise = np.random.normal(0, std, img.shape)
        img = np.clip(img + noise, 0, 255).astype(np.uint8)
        sample["obs"] = img
    return noisy
```

```python
def add_occlusion(data, size=8):
    noisy = deepcopy(data)
    for sample in noisy:
        img = sample["obs"].copy()
        H, W, _ = img.shape
        x = np.random.randint(0, W - size)
        y = np.random.randint(0, H - size)
        img[y:y+size, x:x+size] = 0
        sample["obs"] = img
    return noisy
```

```python
def drop_frames(data, drop_rate=0.1):
    noisy = []
    for s in data:
        if random.random() > drop_rate:
            noisy.append(deepcopy(s))
    return noisy
```

```python
def shuffle_blocks(data, block_size=10):
    noisy = deepcopy(data)
    blocks = []
    current = []

    for s in noisy:
        current.append(s)
        if len(current) == block_size:
            blocks.append(current)
            current = []
    if len(current) > 0:
        blocks.append(current)

    random.shuffle(blocks)

    flattened = []
    for b in blocks:
        flattened.extend(b)
    return flattened
```

```python
noisy_data = expert_data

# thêm nhiễu hành động
noisy_data = add_action_flip_noise(noisy_data, p=0.15)

# chèn random hành động
noisy_data = add_random_action_noise(noisy_data, p=0.1)

# nhiễu Gaussian
noisy_data = add_gaussian_noise(noisy_data, std=8)

# occlusion
noisy_data = add_occlusion(noisy_data, size=6)

# drop frames
noisy_data = drop_frames(noisy_data, drop_rate=0.05)

# shuffle block
noisy_data = shuffle_blocks(noisy_data, block_size=12)

print("Final noisy dataset:", len(noisy_data))
```

```python
# -----------------------------
# CHUNKING
# -----------------------------
def chunk_traj(traj, chunk_size=25):
    return [traj[i:i + chunk_size] for i in range(0, len(traj), chunk_size)]


def encode_obs(image_np):
    return base64.b64encode(image_np.tobytes()).decode("utf-8")


def build_chunk_payload(chunk):
    return [
        {
            "obs_b64": encode_obs(step["obs"]),
            "action": step["action"]
        }
        for step in chunk
    ]
```

```python
# -----------------------------
# PROMPT
# -----------------------------
PROMPT_TEMPLATE = """
You are cleaning corrupted demonstration data from the MiniGrid environment.

You receive a SEQUENCE of steps.
Each step contains:
- obs_b64: base64-encoded RGB image (8x8)
- action: integer (0..6)

Your task:
1. Infer the intended movement direction.
2. Detect noisy/inconsistent wrong actions.
3. Correct them to produce a smooth, consistent path.
4. Output ONLY a JSON list of integers.

Example output:
[0, 0, 1, 1, 2, 2]
"""
```

```python
# -----------------------------
# CALL GEMINI WITH RETRY
# -----------------------------
def call_gemini_with_retry(prompt_text, max_retries=4, delay=4):
    for attempt in range(max_retries):
        try:
            response = model.generate_content(
                contents=[
                    {"parts": [{"text": prompt_text}]}  # <-- đúng format
                ]
            )
            return response.text
        except Exception as e:
            print(f'Transient error calling model (attempt {attempt+1}/{max_retries}): "{e}"')
            time.sleep(delay)

    print("FATAL: All retries failed. Returning None.")
    return None
```

```python
def filter_chunk_with_gemini(chunk):
    payload = build_chunk_payload(chunk)

    final_prompt = PROMPT_TEMPLATE + "\n\n" + json.dumps(payload)

    raw_text = call_gemini_with_retry(final_prompt)
    if raw_text is None:
        print("ERROR: Gemini returned None. Using original actions.")
        return [s["action"] for s in chunk]

    # parse JSON safely
    try:
        cleaned_actions = json.loads(raw_text)
        if isinstance(cleaned_actions, list) and len(cleaned_actions) == len(chunk):
            return cleaned_actions
        else:
            raise ValueError("Bad JSON format")
    except:
        print("ERROR PARSING RESPONSE, returning original actions.")
        return [s["action"] for s in chunk]
```
